# Feature Matrix Construction

This notebook describes the transformation of correlation maps into
feature vectors suitable for machine learning classifiers.


```python
import numpy as np
import matplotlib.pyplot as plt
```

## Feature Extraction Overview

Correlation maps encode spatial relationships between microphone channels.
To preserve these relationships while enabling classification, the maps
are converted into structured feature vectors.


```python
corr_map = np.random.rand(4, 12)
print("Correlation map shape:", corr_map.shape)
```

## Feature Vector Construction

Based on experimental observations, the first row of the correlation map
contains the most discriminative information. Therefore, only this row
is selected to form the feature vector.


```python
feature_vector = corr_map[0, :]
print("Feature vector shape:", feature_vector.shape)
```

```python
num_samples = 120
feature_matrix = np.array([
    np.random.rand(12) for _ in range(num_samples)
])

print("Feature matrix shape:", feature_matrix.shape)
```

```python
plt.figure(figsize=(8, 4))
plt.boxplot(feature_matrix)
plt.title("Feature Distribution Across Samples")
plt.xlabel("Feature Index")
plt.ylabel("Value")
plt.show()
```

## Summary

This notebook demonstrated the conversion of correlation maps into
feature vectors by selecting row-wise correlation patterns.
The resulting feature matrix will be used for training and evaluating
classification models in the next stage.
