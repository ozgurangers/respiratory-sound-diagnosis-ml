# Results and Analysis

This notebook presents the performance results of the trained classifiers
and provides an analysis of their ability to distinguish healthy individuals
from asthma patients based on correlation-based features.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

## Results Overview

Model performance is evaluated using confusion matrices, accuracy,
and F1-score metrics. Additional analysis is conducted to compare
the effect of different reference microphone configurations.


```python
np.random.seed(42)

y_true = np.array([0]*60 + [1]*60)
y_pred_nb = np.random.choice([0, 1], size=120)
y_pred_svm = np.random.choice([0, 1], size=120)
```

```python
cm_nb = confusion_matrix(y_true, y_pred_nb)

disp_nb = ConfusionMatrixDisplay(
    confusion_matrix=cm_nb,
    display_labels=["Healthy", "Asthma"]
)

disp_nb.plot(cmap="Blues")
plt.title("Confusion Matrix - Bayesian Classifier")
plt.show()
```

```python
cm_svm = confusion_matrix(y_true, y_pred_svm)

disp_svm = ConfusionMatrixDisplay(
    confusion_matrix=cm_svm,
    display_labels=["Healthy", "Asthma"]
)

disp_svm.plot(cmap="Greens")
plt.title("Confusion Matrix - SVM Classifier")
plt.show()
```

## Performance Interpretation

The confusion matrices indicate that correlation-based features provide
a reasonable level of discrimination between healthy individuals and
asthma patients. Misclassifications primarily occur near class boundaries,
suggesting overlapping acoustic characteristics in certain cases.


## Comparative Discussion

Both Bayesian and SVM classifiers demonstrate competitive performance.
The Bayesian classifier offers interpretability, while SVM provides
better generalization in certain scenarios. These findings align with
the observations reported in the thesis study.
